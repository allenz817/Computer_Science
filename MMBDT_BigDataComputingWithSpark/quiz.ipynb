```python
from pyspark.sql import SparkSession
from pyspark import SparkContext

sc = SparkContext.getOrCreate()
spark = SparkSession(sc)

A = sc.parallelize(range(1, 1000))
t = 50
B = A.filter(lambda x: x < t)
count_B = B.count()
B.cache() 
# ensures that B's computed results (numbers 1-49) are stored in memory, so when C is created, 
    # it uses the cached data rather than re-evaluating B's transformation with the new value of t.
t = 10
C = B.filter(lambda x: x > t)
count_C = C.count()

result = [count_B, count_C]

print(result)
```
