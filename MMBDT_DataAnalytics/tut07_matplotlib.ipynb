### Other Formatting

Grid lines
- [matplotlib.pyplot.grid()](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.grid.html#matplotlib.pyplot.grid)
- default to toggle the visibility of the grid lines
- plt.grid(True) to make gridlines visisble, plt.grid(False) otherwise

```python
x = [1, 2, 3, 4]
y1 = [1.5, 4, 2.2, 3.3]
y2 = [2, 3, 4, 5]
plt.plot(x, y1);  # Plot with default settings
plt.plot(x, y2);  # Plot with blue solid lines and triangle markers

plt.ylabel('Y Axis')

plt.xlabel('X Axis')

plt.title('A Simple Example') 

plt.grid()

plt.figlegend(['green line', 'blue line'])

# note that the order of legend labels corresponds to the order of plot calls
```

### Plotting with Basics Formatting Using Format Strings

#### Examples:
'b' --  blue markers with default shape  
'or' -- red circles  
'-g' -- green solid line  
'--' -- dashed line with default color  
'^k:' -- black triangle_up markers connected by a dotted line  

#### Markers
|Character|Description|Character|Description|Character|Description|Character|Description|
|---|---|---|---|---|---|---|---|
|'.'|point marker|','|pixel marker|'o'|circle marker||
|'v'|triangle_down marker|'^'|triangle_up marker|'<'|triangle_left marker|'>'|triangle_right marker|
|'1'|tri_down marker|'2'|tri_up marker|'3'|tri_left marker|'4'|tri_right marker|
|'8'|octagon marker|'s'|square marker|'p'|pentagon marker|'P'|plus (filled) marker|
|'*'|star marker|'h'|hexagon1 marker|'H'|hexagon2 marker|'+'|plus marker|
|'x'|x marker|'X'|x (filled) marker|'D'|diamond marker|'d'|thin_diamond marker|
|'\|'|vline marker|'_'|hline marker|

#### Line Styles
|Character|Description|Character|Description|
|---|---|---|---|
|'-'|solid line style|'--'|dashed line style|
|'-.'|dash-dot line style|':'|dotted line style|


#### Colors

The supported color abbreviations are the single letter codes:

|Character|Color|Character|Color|Character|Color|Character|Color|
|---|---|---|---|---|---|---|---|
|'b'|blue|'g'|green|'r'|red|'c'|cyan|
|'m'|magenta|'y'|yellow|'k'|black|'w'|white|


```python
x = [1, 2, 3, 4]
y1 = [1.5, 4, 2.2, 3.3]
y2 = [2, 3, 4, 5]
plt.plot(x, y1, '--^g')  # Plot with dashed lines and triangle markers in green
plt.plot(x, y2, '-or')  # Plot with solid lines and circle markers in red

plt.title('Using Format Strings')
```

Plot y versus x

- [Axes.set_title](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_title.html#matplotlib.axes.Axes.set_title)

- [Axes.set_xlabel](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_xlabel.html#matplotlib.axes.Axes.set_xlabel)

Set the label for the y-axis
- [Axes.set_ylabel](https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_ylabel.html#matplotlib.axes.Axes.set_ylabel)

Set the current tick locations and labels of the x-axis
- [matplotlib.pyplot.xticks](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xticks.html)
- e.g. xticks(x-coordinates, labels, font size, rotation angle)

Set the current tick locations and labels of the y-axis
- [matplotlib.pyplot.yticks](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.yticks.html)
- e.g. yticks(y-coordinates, labels, font size, rotation angle)


```python
fig, ax = plt.subplots()  # Create a figure containing a single axes.

x = [1, 2, 3, 4]
y = [1.5, 4, 2.2, 3.3]

ax.plot(x, y, '-.cD') # Plot with dash-dot lines and diamond markers in cyan

ax.set_ylabel('y-axis')
ax.set_xlabel('x-axis')

ax.set_title('A subplot') # Set a title for the axes.

fig.suptitle("A Figure") # Set a title for the figure.

fig.legend(["line 1"])
```

```python
fig, axs = plt.subplots(2, 2, figsize=(10,7.5))  # Create a figure containing a single axes.
x = [1, 2, 3, 4, 5]
y1 = [1.5, 4, 2.2, 3.3, 0.2]
y2 = [2, 3, 4, 5, 8]
y3 = [92, 22, 45, 2, 78]
y4 = [23, 34, 45, 56, 67]

axs[0, 0].plot(x, y1, '--^g');  # top-left plot with dashed lines and triangle markers in green
axs[0, 1].plot(x, y2, '-or');  # top-right plot with solid lines and circle markers in red
axs[1, 0].plot(x, y3, '-.Db');  # bottom-left plot with dashed lines and triangle markers in green
axs[1, 1].plot(x, y4, ':sm');  # bottom-right plot with solid lines and circle markers in red
```

## Other Types of Plots

#### Bar Chart
- [matplotlib.pyplot.bar](https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.pyplot.bar.html)
- For example: 
  - plt.bar(x, heights, align='center')  
    where x is x-coordinates, heights is the heights of the bars, and center the base of the bars

#### Pie Chart 
- [matplotlib.pyplot.pie](https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.pyplot.pie.html)
- The wedges are plotted counterclockwise, by default starting from the x-axis.
- For example:
  - plt.pie(x, labels, autopct='%1.1f%%')  
    Make a pie chart of array x, with labels formatted with the corresponding numeric values as percentages

#### Scatter Plot
- [matplotlib.pyplot.scatter](https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.pyplot.scatter.html)
- For example:  
  - plt.scatter(x,y)  
    Plot x versus y

## Plot a Bar Chart

Make a bar plot
- [matplotlib.pyplot.bar](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html#matplotlib.pyplot.bar)

```python
# Create a dictionary
table_data = { 'Agriculture': 3.4, 
               'Construction': 12.1, 
               'Manufacturing': 16.2, 
               'Transportation & Utilities': 9.8, 
               'Trade': 19.3, 'Services': 24.3,
               'Government': 14.9}

# Set title
fig, ax = plt.subplots()
ax.set_title('Percent of Total')

# bar(x coordinates, height of bars, alignment)
plt.bar(range(len(table_data)), list(table_data.values()), align='center')

# xticks(x coordinates, labels, font size, rotation angle)
plt.xticks(range(len(table_data)), list(table_data.keys()), fontsize=8, rotation=50)

# Show the barchart
# plt.show()
```

## Plot a Pie Chart

Make a pie chart of array x with labels
- [matplotlib.pyplot.pie(x, explode, labels)](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html)


```python
# Create a dictionary
# Create a dictionary
table_data = { 'Agriculture': 3.4, 
               'Construction': 12.1, 
               'Manufacturing': 16.2, 
               'Transportation & Utilities': 9.8, 
               'Trade': 19.3, 'Services': 24.3,
               'Government': 14.9}

# Import numpy
import numpy as np

# pie(data values, explode, labels, format)
plt.pie(list(table_data.values()), labels=list(table_data.keys()), autopct='%1.1f%%')

# Axis ratio
plt.axis('equal')

# Show piechart
#plt.show()
```
